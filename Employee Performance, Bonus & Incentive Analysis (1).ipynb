```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Employee_Data.xlsx")
df.head()
```

```python
df.info()
df.describe()
```

```python
dept_perf = df.groupby("Department")["Performance %"].mean()

dept_perf.plot(kind="bar")
plt.title("Average Performance % by Department")
plt.xlabel("Department")
plt.ylabel("Performance %")
plt.show()
```

```python
dept_salary = df.groupby("Department")["Final Salary"].mean()

dept_salary.plot(kind="bar")
plt.title("Average Final Salary by Department")
plt.xlabel("Department")
plt.ylabel("Final Salary")
plt.show()
```

```python
sns.scatterplot(x="Performance %", y="Bonus", data=df)
plt.title("Performance vs Bonus")
plt.show()
```

```python
sns.countplot(x="Reward Status", data=df)
plt.title("Employee Reward Status Distribution")
plt.show()
```

```python
top10 = df.sort_values("Performance %", ascending=False).head(10)

sns.barplot(x="Name", y="Performance %", data=top10)
plt.xticks(rotation=45)
plt.title("Top 10 Employee Performers")
plt.show()
```
